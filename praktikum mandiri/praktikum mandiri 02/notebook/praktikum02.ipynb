```python
from sklearn.model_selection import train_test_split

# Membagi dataset menjadi Training (80%) dan Testing (20%)
train_df, test_df = train_test_split(df_day, test_size=0.2, random_state=42)

print("Jumlah total data :", len(df_day))
print("Jumlah data Training :", len(train_df))
print("Jumlah data Testing :", len(test_df))
```

```python
# Membagi Training lagi menjadi Training (90%) dan Validation (10%)
train_df, val_df = train_test_split(train_df, test_size=0.1, random_state=42)

print("Jumlah data Training setelah split:", len(train_df))
print("Jumlah data Validation:", len(val_df))
```

```python
# Data Testing: 20% dari total dataset
train_df, test_df = train_test_split(df_day, test_size=0.2, random_state=42)

print("Jumlah data Testing :", len(test_df))
```
